### 2. Check Odds (0.4p)

Implement the function `check_odds(numbers)` that receives a list of positive integers and returns `True` if the count of odd numbers exceeds the count of even numbers. Otherwise (lower or equal), return `False`.

**Example**:
```python
# Example usage

print(output)  
# True (there are 3 odd numbers vs 2 even)
```

**Assumptions**:
- `numbers` is non-empty.
- Each element in `numbers` is a positive integer.

````python
def check_odds(numbers):
  odd_count = even_count = 0
  for number in numbers:
    if number % 2 == 0:
      even_count += 1
    else:
      odd_count += 1
  if odd_count > even_count:
    return True
  else:
    return False
````

### 8. XSort (1.0p)

Implement the function `xsort(numbers)` that transforms a list of integers as follows:

1. Sort the list in **descending** order.
2. For the first element, keep it as is. For each subsequent element, add it to the *previous* (originally sorted) element.
3. Return the resulting list of these sums.

**Example**:
```python
# Example usage
x = xsort([2, 3, 1, 5])
print(x)
# [5, 8, 5, 3]
# Explanation:
# 1) Desc sort -> [5, 3, 2, 1]
# 2) Summation -> [5, (3+5), (2+3), (1+2)] = [5, 8, 5, 3]
```

**Assumptions**:
- `numbers` is non-empty and up to length 10^5.
- Each element is within [-10^5, 10^5].

````python
def xsort(numbers):
  numbers.sort(reverse=True)
  xsorted = []
  for i in range(len(numbers)):
    if i == 0:
      xsorted.append(numbers[i])
    else:
      xsorted.append(numbers[i] + numbers[i-1])

  return xsorted
````

Implement the function `find_duplicates(nums)` that, given a list of integers, **returns a list of the values that appear more than once**, sorted from smallest to largest.

**Example**:
```python
duplicates = find_duplicates([4, 3, 2, 7, 8, 2, 3, 1])
print(duplicates)
# [2, 3]
```
- Both 2 and 3 appear more than once, hence they‚Äôre returned, sorted as `[2, 3]`.

**Assumptions**:
- `nums` is non-empty and has length at most 10^5.
- Each element of `nums` is an integer in the range [-10^5, 10^5].

````python
def find_duplicates(nums):
  nums1 = []
  duplicates = []
  for n in nums:
    if n not in nums1:
      nums1.append(n)
    elif n not in duplicates:
      duplicates.append(n)
  duplicates.sort()
  return duplicates
````

Implement the function `ruin_string(my_string)` that removes **duplicate characters** from the input (keeping only the first occurrence of each character), and then **reverses** the resulting string.

**Example**:
```python

result2 = ruin_string("The answer to the ultimate question of life, the universe, and everything is 42.")
print(result2)
# ".24gydv,fqmiluotrwsna ehT"
```

**Assumptions**:
- `my_string` is a non-empty string.
- It consists of ASCII characters (ordinal values 0‚Äì127).
- Uppercase and lowercase characters are considered distinct (`'A'` vs `'a'`).

````python
def ruin_string(my_string):
  str_chars = []
  for char in list(my_string):
    if char not in str_chars:
      str_chars.append(char)
  str_chars.reverse()
  return "".join(str_chars)
````

## Part II - Building KMeans from scratch (12.0p)

**K-Means** is one of the most widely-used clustering algorithms in data analysis and machine learning. Its primary goal is to divide a dataset into \( k \) groups (called *clusters*) such that each group of points is centered around a location we call a *centroid*. A point is considered part of the cluster whose centroid it is closest to (often using Euclidean distance).

Here‚Äôs a high-level overview of the K-Means procedure:

1. **Choose \( k \) (number of clusters):** This determines how many centroids we need to find.

2. **Initialize the centroids:** A common approach is to randomly select \( k \) distinct points from the dataset to serve as the initial centroid positions.

3. **Measure distances and assign clusters:** For each data point, compute its distance to each of the \( k \) centroids, then assign the point to the cluster with the closest centroid.

1. **Recompute centroid positions:** Once each point is assigned, update each centroid‚Äôs location by taking the average of all points that belong to its cluster.

2. **Check for convergence:** Measure how much the centroids have moved compared to their previous positions (or compare how the total within-cluster distance‚Äîoften called *inertia*‚Äîhas changed). If the centroid shifts are below a certain *tolerance* or if the change in inertia is small, we consider the solution to be stable.

3. **Repeat until stable or maximum iterations reached:** Reassign points (Step 3), update centroids (Step 4), and check convergence (Step 5) until either the movement of centroids is sufficiently small or you‚Äôve hit the maximum number of iterations (`max_iter`).

At the end of this iterative process, you‚Äôll have:

- A final set of \( k \) centroids
- Each data point assigned to exactly one of those \( k \) clusters

While K-Means is guaranteed to converge (i.e., stop updating the centroids after enough iterations), it might reach different solutions depending on how the centroids were initially placed. In practice, it‚Äôs common to run K-Means multiple times and pick the best result based on an evaluation metric like inertia (the sum of squared distances between points and their assigned centroids).

### 1. Distance Function (1.0p)

Implement a function `euclidean_distance(a, b)` that computes the Euclidean distance between two vectors (lists or tuples) `a` and `b`. **No external libraries (e.g., NumPy) are allowed.**

\begin{equation}
    d(a, b) = \sqrt{\sum_{i=0}^{D} (xi-yi)^2}
\end{equation}

The function should return the distance as a floating-point number.

**Example 1: 2D**

```python

```

````python
def euclidean_distance(a, b):
    """
    Compute and return the Euclidean distance between vectors a and b.

    Parameters:
    -----------
    a : list or tuple
        Coordinates of the first point
    b : list or tuple
        Coordinates of the second point

    Returns:
    --------
    float
        The Euclidean distance between a and b
    """
    diff_roots = []
    for i in range(len(a)):
      diff_roots.append((b[i] - a[i])**2)

    return math.sqrt(sum(diff_roots))
````

### 4. Update Centroids (2.0p)

Implement a function `update_centroids(X, labels, k)` that recalculates the centroid positions based on the current cluster assignments.

- For each cluster index `i` (from `0` to `k-1`), compute the mean of all points belonging to that cluster.

**Example:**

```python
# Input:
X = [[1, 2], [2, 3], [10, 10], [9, 9]]
labels = [0, 0, 1, 1]

new_centroids = update_centroids(X, labels, k)
print(new_centroids)
# Expected output: [[1.5, 2.5], [9.5, 9.5]]

# Why?
# Cluster 0: points [1, 2] and [2, 3]
#   - Mean of x-coordinates: (1 + 2) / 2 = 1.5
#   - Mean of y-coordinates: (2 + 3) / 2 = 2.5
#   ‚Üí New centroid for cluster 0: [1.5, 2.5]

# Cluster 1: points [10, 10] and [9, 9]
#   - Mean of x-coordinates: (10 + 9) / 2 = 9.5
#   - Mean of y-coordinates: (10 + 9) / 2 = 9.5
#   ‚Üí New centroid for cluster 1: [9.5, 9.5]
```


````python
def update_centroids(X, labels, k):
    """
    Given the data points X and their assigned cluster labels,
    compute the new centroid coordinates for each cluster.

    Parameters:
    -----------
    X : list of lists/tuples
        Dataset where each row is a sample
    labels : list of int
        The index of the cluster each data point belongs to
    k : int
        Number of clusters

    Returns:
    --------
    new_centroids : list of lists/tuples
        The updated centroid positions
    """
    kXs = []
    for k_index in range(k):
      Xs = []
      for label_index in range(len(labels)):
        if k_index == labels[label_index]:
          # fetch coordinate
          Xs.append(X[label_index])
      kXs.append(Xs)

    updated_centroids = []
    axes = len(kXs[0][0])
    for kX in kXs:
      mean_axis = []
      for i in range(axes):
        sum = 0
        for X in kX:
          sum += X[i]
        mean_axis.append(sum/len(kX))
      updated_centroids.append(mean_axis)
    return updated_centroids
````

````python
# Input:
X = [[1, 2], [2, 3], [10, 10], [9, 9]]
labels = [0, 0, 1, 1]
k = 2

new_centroids = update_centroids(X, labels, k)
print(new_centroids)
# Expected output: [[1.5, 2.5], [9.5, 9.5]]
````

### 5. Compute Inertia (1.5p)

Implement a function `compute_inertia(X, labels, centroids)` that returns the sum of squared distances of each point to its assigned centroid. This is often used as a measure of how tight the clusters are.

$$
\text{Inertia} = \sum_{i=1}^{n} \left\| x_i - \mu_{c_i} \right\|^2
$$

Where:

- $n$ is the number of data points,  
- $x_i$ is the $i$-th data point,  
- $\mu_{c_i}$ is the centroid of the cluster assigned to $x_i$,  
- $| \cdot |^2$ is the squared Euclidean distance.

**üí° HINT:**
- Use the euclidean distance from the previous exercise.
- For each data point, find the centroid of the cluster it was assigned to.
- Then, compute the **squared Euclidean distance** between the point and its assigned centroid.
- Do this for every point in the dataset.
- Add up all the squared distances ‚Äî this total is the inertia.

**Example:**

```python
# Input:
X = [[1,2], [2,3], [10,10], [9,9]]
labels = [0, 0, 1, 1]
centroids = [[1.5, 2.5], [9.5, 9.5]]

inertia_val = compute_inertia(X, labels, centroids)
print(inertia_val)
# Expected output: ~2.0
```

````python
def compute_inertia(X, labels, centroids):
    """
    Compute the sum of squared distances of points to their assigned centroids.

    Parameters:
    -----------
    X : list of lists
        Dataset where each row is a sample
    labels : list of int
        The index of the cluster each data point belongs to
    centroids : list of lists/tuples
        The current centroid positions

    Returns:
    --------
    float
        The sum of squared distances
    """
    inertia = 0
    for i in range(len(X)):
      data_point = X[i]
      centroid = centroids[labels[i]]
      inertia += euclidean_distance(centroid, data_point)**2
    return inertia
````

````python
class KMeans:

    def __init__(self, n_clusters=2, max_iter=100, tol=0.0001, random_state=None):
        self.n_clusters = n_clusters
        self.max_iter = max_iter
        self.tol = tol
        self.random_state = random_state

        # internal variables needed to track state of centroids, inertia, no of iterations and labels
        self.centroids = None
        self.inertia = None
        self.n_iter = 0
        self.labels_ = None

    def fit(self, X):
        self.centroids = self._init_centroids(X)

        for _ in range(self.max_iter):
          labels = self._assign_clusters(X, self.centroids)
          new_centroids = self._update_centroids(X, labels)
          new_inertia = self._compute_inertia(X, labels, new_centroids)
          self.n_iter += 1

          if self.inertia is not None and abs(self.inertia - new_inertia) < self.tol:
              break

          self.centroids = new_centroids
          self.inertia = new_inertia
          self.labels_ = labels

        pass  # <-- Your code here

    def predict(self, X):
        labels = self._assign_clusters(X, self.centroids)
        return labels

    def _euclidean_distance(self, a, b):
        diff_roots = []
        for i in range(len(a)):
          diff_roots.append((b[i] - a[i])**2)

        return math.sqrt(sum(diff_roots))

    def _assign_clusters(self, X, centroids):
        labels = []
        for data_point in X:
          mini = 0
          mini_index = 0
          for i in range(len(centroids)):
            if i == 0:
              mini = abs(self._euclidean_distance(data_point, centroids[i]))
            elif abs(self._euclidean_distance(data_point, centroids[i])) < mini:
              mini = abs(self._euclidean_distance(data_point, centroids[i]))
              mini_index = i
          labels.append(mini_index)
        return labels

    def _init_centroids(self, X):
        if self.random_state is not None:
          random.seed(self.random_state)
        return random.sample(X, self.n_clusters)

    def _update_centroids(self, X, labels):
        kXs = []
        for k_index in range(self.n_clusters):
          Xs = []
          for label_index in range(len(labels)):
            if k_index == labels[label_index]:
              # fetch coordinate
              Xs.append(X[label_index])
          kXs.append(Xs)

        updated_centroids = []
        axes = len(kXs[0][0])
        for kX in kXs:
          mean_axis = []
          for i in range(axes):
            sum = 0
            for X in kX:
              sum += X[i]
            mean_axis.append(sum/len(kX))
          updated_centroids.append(mean_axis)
        return updated_centroids

    def _compute_inertia(self, X, labels, centroids):
        inertia = 0
        for i in range(len(X)):
          data_point = X[i]
          centroid = centroids[labels[i]]
          inertia += self._euclidean_distance(centroid, data_point)**2
        return inertia
        pass  # <-- Your code here
````
